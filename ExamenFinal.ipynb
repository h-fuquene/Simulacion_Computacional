Para los siguientes generadores congruenciales mixtos encuentre $u_1,...,u_{100}$, e indique el tamaño del ciclo (si lo tiene).

**b)** $x_0=3, x_n = (5x_{n-1}+7) \mod 200$

**d)** $x_0=7, x_n = (5x_{n-1}+3) \mod 33$

**e)** $x_0=1733, x_n = (7x_{n-1}) \mod 144$


```python
from datetime import datetime
import random

def metodo_congruencial_mixto(cantidad=1, x0=None, a=None, b=None, mod=None):
  pseudoaletorios = []
  if x0 !='val':
      x0
  if a != 'val':
      a
  if b !='val':
      b
  if mod !='val':
      mod 
  actual=x0
  for i in range(0,cantidad):
      xn=(a*actual+b)%mod
      un=xn/mod
      pseudoaletorios.append(un)
      actual=xn
  return pseudoaletorios
def calcular_periodo(lista_aleatorios):
    periodo=0
    for i in range(0, len(lista_aleatorios)):
      for j in range(i+1, len(lista_aleatorios)):
          if lista_aleatorios[i]==lista_aleatorios[j]:
              periodo=j-i
              break
      if periodo !=0:
          break

    return periodo
```

Para la generación de las variables aleatorias de los tiempos de atención del servidor (variable aleatoria $Y$) debe generar variables aleatorias continuas de una distribución exponencial con $\lambda = B$ a partir de un generador congruencial Turbo Pascal (i.e. $a = 134775813$, $c = 1$, $m=2^{32}$) y semilla $X_0 = 123456789$.

```python
%matplotlib inline 

import numpy as np
import math
import matplotlib.pyplot as plt

Xant1 = 987654321
Xant2 = 123456789

M = 99999.0
T = 100.0
at = []
an = []
LLt = []
St = []

def gencongru1(): 
    global Xant1
    a = 22695477
    c = 1
    m = 2 ** 32
    Xnext = (a * Xant1 + c) % m
    Xant1 = Xnext
    U = Xnext / m
    return U

def gencongru2():
    global Xant2
    a = 134775813
    c = 1
    m = 2 ** 32
    Xnext = (a * Xant2 + c) % m
    Xant2 = Xnext
    U = Xnext / m
    return U

def GenerarX():
    vlambda = 3
    U = gencongru1()
    X = -(math.log(1-U))/float(vlambda)
    return X

def GenerarY():
    vlambda = 7
    U = gencongru2()
    Y = -(math.log(1-U))/float(vlambda)
    return Y

def Llegada(tsuc):
    global n, NLL, NS, t, S, Serv, LL, at, an
    t=tsuc
    n = n + 1
    LLt.append(t)
    at.append(t)
    an.append(n)
    
    NLL = NLL + 1
    LL.append(t)
    
    X = GenerarX()
    
    if (t+X) < T:
        TSuc['tLL'] = t + X
    if n==1:
        Y = GenerarY()
        TSuc['tS'] = t + Y
        #Serv[NS]=Y 
        Serv.append(Y)
        
def Servidor(tsuc):
    global n, NLL, NS, t, S, Serv, LL, at, an
    t = tsuc
    n = n - 1
    
    St.append(t)
    at.append(t)
    an.append(n)
    
    NS = NS + 1
    S.append(t)

    if n>0:
        Y = GenerarY()
        TSuc['tS'] = t + Y
        Serv.append(Y) 
    
t = tsuc = NLL = NS = n = 0

at.append(t)
an.append(n)

TSuc = {"tLL":M, "tS":M}

LL = []
S = []
Serv = []
LL.append(0.0)
S.append(0.0)
Serv.append(0.0)

X = GenerarX()

if X > T:
    Tp = t_med_sistema = t_med_cola = 0.0
    exit
else:
    Llegada(X)
    while((TSuc['tLL'] != M)|(TSuc['tS']!=M)):
        if TSuc['tLL'] < TSuc['tS']:
            tsuc = TSuc['tLL']
            TSuc['tLL'] = M
            Llegada(tsuc)
            
        if TSuc['tS'] < TSuc['tLL']:
            tsuc = TSuc['tS']
            TSuc['tS'] = M
            Servidor(tsuc)

    Tp = max(0,t-T)
    acumulo1 = acumulo2 = 0.0
    ind = 0
    
    while ind < NLL:
        acumulo1 = acumulo1 + S[ind] - LL[ind]
        acumulo2 = acumulo2 + S[ind] - LL[ind] - Serv[ind]
        ind = ind + 1
    t_med_sistema = acumulo1/NLL
    t_med_cola = acumulo2/NLL

    exit
```

```python
from random import randint
import simpy 
```

```python
TALKS_PER_SESSION = 3
TALK_LENGTH = 20
BREAK_LENGTH = 30
DURATION_EAT = 2
BUFFET_SLOTS = 20
```

```python
def attendee(env, name, buffet, knowledge=0, hunger=0):
  global hunger_total
  global knowledge_total
  
  while True:
    # Visit talks
    for i in range(TALKS_PER_SESSION):
      knowledge = knowledge + randint(0, 3) / (1 + hunger)
      hunger = hunger + randint(1, 4)
      yield env.timeout(TALK_LENGTH)
    print('El asistente %s terminó las charlas con un conocimiento %.2f y hambre '
    '%.2f.' % (name, knowledge, hunger))

    # Go to buffet
    start = env.now
    with buffet.request() as req:
      yield req | env.timeout(BREAK_LENGTH - DURATION_EAT)
      time_left = BREAK_LENGTH - (env.now - start)
      
      if req.triggered:
        food = min(randint(3, 12), time_left) # Less time -> less food
        yield env.timeout(DURATION_EAT)
        hunger -= min(food, hunger)
        time_left -= DURATION_EAT
        print('El asistente %s terminó de comer con hambre %.2f' %
              (name, hunger))
        hunger_total[name] = hunger
        knowledge_total[name] = knowledge
      else:
        hunger += 1 # Penalty for only taking a look at all the food.
        print('El asistente %s no llegó al buffet :( , el hambre es ahora '
        'at %.2f.' % (name, hunger))
        hunger_total[name] = hunger
        knowledge_total[name] = knowledge
    yield env.timeout(time_left)
```

```python
env = simpy.Environment()
buffet = simpy.Resource(env, capacity=BUFFET_SLOTS)

hunger_total = np.zeros(100)
knowledge_total = np.zeros(100)

for i in range(100):
  env.process(attendee(env, i, buffet))
env.run(until=270)

hunger_sum = sum(hunger_total)
knowledge_sum = sum(knowledge_total)

print()
```
